# Data Visualization

```python
for column in df.columns:
        fig, ax = plt.subplots()
        ax.bar(df[column].value_counts().index, df[column].value_counts().values)
        ax.set_xlabel(column)
        ax.set_ylabel('Frequency')
        plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(data.data, data.target, test_size=0.3, random_state=0)
```

## Train Model using Entropy criterion

```python
dtc_model = DecisionTreeClassifier(criterion="entropy", max_depth=4, random_state=10)
dtc_model.fit(X_train, y_train)
DecisionTreeClassifier(criterion='entropy')
```

```python
y_pred = dtc_model.predict(X_test)
accuracy_entropy = accuracy_score(y_test, y_pred)
print(f"Decision tree entropy accuracy: {accuracy_entropy:.2f}")
```

## Train Model using GINI

```python
dtc_gini = DecisionTreeClassifier(criterion="gini")
dtc_gini.fit(X_train, y_train)
```

```python
y_pred = dtc_gini.predict(X_test)
accuracy_gini = accuracy_score(y_test, y_pred)
print(f"Decision tree gini accuracy: {accuracy_gini:.2f}")
```

# Compare performance

```python
print(f"Decision Tree Classifier using Entropy has an accuracy:{accuracy_entropy*100}%")
print(f"Decision Tree Classifier using Gini has an accuracy:{accuracy_gini*100}%")
```
